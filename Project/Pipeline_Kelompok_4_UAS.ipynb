# **INI FUNGSI GABUNGAN (TIDAK DI RUN)**

```python
#INI FUNGSI GABUNGAN GAK DI RUN
def transformasi():
    # baca data
    dfHDI = pd.read_csv('data_HDI.csv')
    dfGDP = pd.read_csv('data_GDP.csv')
    dfMentalHealth = pd.read_csv('data_Kewarasan.csv')
    dfKematian = pd.read_csv('data_Kematian.csv')

    # merge
    df_gg = pd.merge(dfHDI, dfMentalHealth, on=['Year', 'Entity'], suffixes=('_HDI', '_MH'))
    df_gg = pd.merge(df_gg, dfGDP, on=['Year', 'Entity'], suffixes=('', '_GDP'))
    df_gg = pd.merge(df_gg, dfKematian, on=['Year', 'Entity'], suffixes=('', '_Kematian'))

    # kesimpulan
    df_agg = df_gg.groupby('Entity').agg(
        Tahun = ('Year', list),
        Jumlah_Kematian = ('suicides_no', 'mean'),
        Populasi = ('population', 'mean')
    ).reset_index()

    # rename kolom Entity jadi Negara
    df_agg.rename(columns={'Entity': 'Negara'}, inplace=True)

    return df_agg
```

```python
# jalankan fungsi
df_agg = transformasi(dfHDI)
df_agg
```

```python
# 1. fungsi untuk membaca data

def read_data(df_agg):
    df = pd.read_csv(df_agg)
    return df
```

```python
# 4 Fungsi untung menggabungkan 2 data
from functools import reduce
import pandas as pd

def merge_multiple_dfs(dfs, on_cols):
    return reduce(lambda left, right: pd.merge(left, right, on=on_cols), dfs)

# DataFrame punya kolom 'Entity' dan 'Year'
dfs = [df_agg]
df_merged = merge_multiple_dfs(dfs, on_cols=['Negara', 'Tahun'])
```

# Pipeline
Buat fungsi Fungsi Pipeline untuk menjalankan Fungsi Pembersihan dan Transformasi
1. pipeline baca data
2. pipeline hapus baris null
3. pipeline misssing values
4. pipeline merge data

## **PIPELINE**

```python
# Baca Data untuk Pipeline
import pandas as pd
import logging

logging.basicConfig(level=logging.ERROR)

def read_data(df_path):
    try:
        data = pd.read_csv(df_path)
        print(f"Berhasil membaca: {df_path}")
        return data
    except Exception as e:
        print(f"Gagal membaca: {df_path} — {e}")
        return pd.DataFrame()

def pipeline_baca_data(dfHDI, dfGDP, dfKematian, dfMentalHealth):
    try:
        df_HDI = read_data(dfHDI)
        df_GDP = read_data(dfGDP)
        df_Kematian = read_data(dfKematian)
        df_MentalHealth = read_data(dfMentalHealth)
    except Exception as e:
        logging.error(f'ERORR WOIIII: {e}')
        return pd.DataFrame(), pd.DataFrame(), pd.DataFrame(), pd.DataFrame()

    return df_HDI, df_GDP, df_Kematian, df_MentalHealth
```

```python
# pipeline hapus baris null

def pipeline_hapus_baris_null(dfHDI, dfGDP, dfKematian, dfMentalHealth):
    df_a = pd.DataFrame()
    df_b = pd.DataFrame()
    df_c = pd.DataFrame()
    df_d = pd.DataFrame()
    try:
        df_a = drop_missing_values(dfHDI)
        df_b = dfGDP.dropna(axis=1, how='all')
        df_c = drop_missing_values(dfKematian)
        df_d = drop_missing_values(dfMentalHealth)
    except Exception as e:
        logging.error('Data gagal dihapus')
    return df_a, df_b, df_c, df_d
```

```python
# Isi Missing Value
import pandas as pd
import logging

logging.basicConfig(level=logging.ERROR)

def fill_missing_values(df):
    try:
        df = df.fillna(df.mean(numeric_only=True))
    except Exception as e:
        logging.error(f"Gagal mengisi missing values: {e}")
    return df

def pipeline_isi_mv(dfHDI, dfGDP, dfKematian, dfMentalHealth):
    try:
        df_a = fill_missing_values(dfHDI)
        df_b = fill_missing_values(dfGDP)
        df_c = fill_missing_values(dfKematian)
        df_d = fill_missing_values(dfMentalHealth)
    except Exception as e:
        logging.error(f'Data gagal diisi: {e}')
        return pd.DataFrame(), pd.DataFrame(), pd.DataFrame(), pd.DataFrame()

    return df_a, df_b, df_c, df_d
```

```python
# Mengubah nama kolom dan Gabungkan data dengan kolom year dan entity
def merge_data(df1, df2):
    return pd.merge(df1, df2, on=['Tahun', 'Negara'], how='inner')

def pipeline_data(dfHDI, dfGDP, dfKematian, dfMentalHealth):
    try:
        # Rename kolom
        dfHDI = dfHDI.rename(columns={'Year': 'Tahun', 'Entity': 'Negara'})
        dfGDP = dfGDP.rename(columns={'Year': 'Tahun', 'Entity': 'Negara'})
        dfKematian = dfKematian.rename(columns={'Year': 'Tahun', 'Entity': 'Negara', 'Suicides': 'Jumlah_Kematian', 'Population':'Populasi'})
        dfMentalHealth = dfMentalHealth.rename(columns={'Year': 'Tahun', 'Entity': 'Negara'})

        dfKematian_selected = dfKematian[['Tahun', 'Negara', 'Jumlah Kematian', 'Populasi']]
        dfGDP_selected = dfGDP[['Tahun', 'Negara', 'GDP']]
        dfHDI_selected = dfHDI[['Tahun', 'Negara', 'HDI']]

        # merge
        df_merge1 = merge_data(dfHDI, dfMentalHealth)
        df_merge2 = merge_data(df_merge1, dfGDP)
        df_final = merge_data(df_merge2, dfKematian)

    except Exception as e:
        logging.error(f'Data gagal digabungkan: {e}')
        return pd.DataFrame()

    return df_final
```

# Menjalankan Fungsi Pipeline

```python
# 1. Membaca data
print("======== MEMBACA DATA MENTAH ========\n")
df_HDI, df_GDP, df_Kematian, df_MentalHealth = pipeline_baca_data(
    'data_HDI.csv',
    'data_GDP.csv',
    'data_Kematian.csv',
    'data_Kewarasan.csv'
)

display(dfHDI.head())
display(dfGDP.head())
display(dfKematian.head())
display(dfMentalHealth.head())


# 2. Melihat isi missing value
print("\n\n======== MELIHAT ISI MISSING VALUE ========\n\n")
dfHDI, dfGDP, dfKematian, dfMentalHealth = pipeline_isi_mv(
    dfHDI, dfGDP, dfKematian, dfMentalHealth
)

# Tampilkan contoh isi DataFrame setelah isi missing values
display("Preview dfHDI:")
display(dfHDI.head())

display("Preview dfGDP:")
display(dfGDP.head())

display("Preview dfKematian:")
display(dfKematian.head())

display("Preview dfMentalHealth:")
display(dfMentalHealth.head())
```
